```python
#importing the packages
import pandas as pd
import matplotlib.pylab as plt
```

```python
#read the data from the file
df1 = pd.read_csv('df1.txt', sep=" ")
#print the data (just a check)
print(df1)
```

```python
# Q2 : plotting the histogram

#get the column data
Psygroup_data = df1['Psygroup']
# drop the nan values 
modified_data = Psygroup_data.dropna()
#plot the histogram
plt.hist(modified_data)
# Add title and axis names
plt.title('Frequency Distribution of Psygroup')
plt.xlabel('Psygroup')
plt.ylabel('Number of Occurrences')
plt.show()
```

```python
#Q3 : renaming the column
# we can not directly change the column name if id number because of the space in between so we replace space with ( _ )
df1.columns = df1.columns.str.strip().str.replace(' ', '_')
#now rename the column
df1.rename(columns={"id_number":'ID'}, inplace=True)
#print the column names
print(df1.columns)
```

```python
#Q4 : making the cross table
pd.crosstab(df1['Agegroup'],df1['Psygroup'],margins=True)
```

```python
#Q5 finding the duplicate rows by ID
#now lets find the duplicate rows 
# Select all duplicate rows based on one columns
duplicate_rows = df1[df1.duplicated(['ID'],keep=False)]
#printing all the duplicate rows
print(duplicate_rows)

#the number of duplicated rows bu id can be found as 
num_duplicate_rows = duplicate_rows.shape[0]
print("Total number of duplicate rows :"+ str(num_duplicate_rows))
```
